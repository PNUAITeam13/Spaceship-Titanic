```python
features = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'Age', 'VIP']
# List of featured columns
```

```python
X = data_train[features]
y = data_train['Transported']
```

```python
print('Home planets: ', X.HomePlanet.unique())
print('Destination planets: ', X.Destination.unique())
```

```python
X_train_copied = X.copy()
cabins = X_train_copied['Cabin']
#deck/num/side

X_train_copied[['Cabin_Deck', 'Cabin_Num', 'Cabin_Side']] = X_train_copied['Cabin'].str.split('/', expand=True)
X_train_copied = X_train_copied.drop('Cabin', axis='columns')
X_train_copied['Cabin_Num'] = pd.to_numeric(X_train_copied['Cabin_Num'])
X = X_train_copied.copy()
X
```

```python
X[X.columns].describe()
```

```python
X_missing = X.copy()

X_missing['Age'] = X_missing.Age.fillna(X_missing.Age.mean()) #Fill missing Age with average
# Filling the next columns with most frequent values
X_missing['HomePlanet'] = X_missing.HomePlanet.fillna(X_missing.HomePlanet.mode()[0])
X_missing['CryoSleep'] = X_missing.CryoSleep.fillna(X_missing.CryoSleep.mode()[0])
X_missing['Destination'] = X_missing.Destination.fillna(X_missing.Destination.mode()[0])
X_missing['VIP'] = X_missing.VIP.fillna(X_missing.VIP.mode()[0])
```

```python
X = X_missing.copy()
```

```python
from sklearn.preprocessing import OrdinalEncoder
need_to_transform = ['CryoSleep', 'VIP', 'Cabin_Deck', 'Cabin_Side'] #Destination, HomePlanet moved to OneHot

label_X = X.copy()
encoder = OrdinalEncoder()

label_X[need_to_transform] = encoder.fit_transform(label_X[need_to_transform])
```

```python
# Deck & Side - Mode
label_X['Cabin_Deck'] = label_X['Cabin_Deck'].fillna(label_X.Cabin_Deck.mode()[0])
label_X['Cabin_Side'] = label_X['Cabin_Side'].fillna(label_X.Cabin_Side.mode()[0])
```

```python
# --------------------------------------------------
# Deck & Side: Max + 1
# Cabin number - ?
# cabin_deck_max = label_X.Cabin_Deck.max()
# print('Deck Max:', cabin_deck_max)
# label_X['Cabin_Deck'] = label_X['Cabin_Deck'].fillna(cabin_deck_max + 1)

# cabin_side_max = label_X.Cabin_Deck.max()
# print('Side Max:', cabin_side_max)
# label_X['Cabin_Side'] = label_X['Cabin_Side'].fillna(cabin_side_max + 1)
```

```python
# For number: max number
cabin_max_number = label_X.Cabin_Num.max()
print('Cabin max num: ', cabin_max_number)
label_X.Cabin_Num = label_X.Cabin_Num.fillna(cabin_max_number)
```

```python
X = label_X.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder
OH_cols = ['Destination', 'HomePlanet']
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

OH_X_cols = pd.DataFrame(OH_encoder.fit_transform(X[OH_cols]))
OH_X_cols.index = X.index
OH_X_cols.columns = OH_encoder.get_feature_names_out(OH_cols)

num_X = X.drop(OH_cols, axis=1)
OH_X = pd.concat([num_X, OH_X_cols], axis=1)
OH_X
```

```python
X = OH_X.copy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error as mae
estimators = [618]

for estimator in estimators:
    xgboost = XGBRegressor(n_estimators=estimator, random_state=100, learning_rate=0.01)
    xgboost.fit(X_train, y_train)
    predict = xgboost.predict(X_valid)
    print(estimator, mae(y_valid, predict))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
xgbclassifier = XGBClassifier(objective='multi:softmax', num_class=len(set(y_train)), n_estimators=700, learning_rate=0.01)
xgbclassifier.fit(X_train, y_train)

predict = xgbclassifier.predict(X_valid)
print(accuracy_score(y_valid, predict))
```
