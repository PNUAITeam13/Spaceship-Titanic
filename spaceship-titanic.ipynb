```python
features = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'Age', 'VIP']
# List of featured columns
```

```python
X = data_train[features]
y = data_train['Transported']
```

```python
X
```

```python
y
```

```python
from data_transformer.DataTransformer import DataTransformer

transformer = DataTransformer(replace_data_after_transformation=True, use_backups=True)
```

```python
X_train_copied = X.copy()
cabins = X_train_copied['Cabin']
#deck/num/side

X_train_copied[['Cabin_Deck', 'Cabin_Num', 'Cabin_Side']] = X_train_copied['Cabin'].str.split('/', expand=True)
X_train_copied = X_train_copied.drop('Cabin', axis='columns')
X_train_copied['Cabin_Num'] = pd.to_numeric(X_train_copied['Cabin_Num'])
X = X_train_copied.copy()
X
```

```python
X[X.columns].describe()
```

```python
X_missing = X.copy()
X_missing = transformer.fill_na_supplier(X_missing, ['Age'], lambda: X_missing.Age.mean()).get_data() #Fill missing Age with average
# Filling the next columns with most frequent values
X_missing = transformer.fill_na_function(X_missing, 
                                         ['HomePlanet', 'CryoSleep', 'Destination', 'VIP', 'Cabin_Deck', 'Cabin_Side'],
                                         lambda column: column.mode()[0]).get_data()
```

```python
# For number: max number
cabin_max_number = X_missing.Cabin_Num.max()
print('Cabin max num: ', cabin_max_number)
X_missing = transformer.fill_na_supplier(X_missing, ['Cabin_Num'], lambda: cabin_max_number).get_data()
```

```python
X = X_missing.copy()
```

```python
missingno.matrix(X)
```

```python
X = transformer.use_ordinal_encoder(X, ['CryoSleep', 'VIP', 'Cabin_Deck', 'Cabin_Side']).get_data()
```

```python
X = transformer.use_one_hot_encoder(X, ['Destination', 'HomePlanet']).get_data()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error as mae
estimators = [618]

for estimator in estimators:
    xgboost = XGBRegressor(n_estimators=estimator, random_state=100, learning_rate=0.01)
    xgboost.fit(X_train, y_train)
    predict = xgboost.predict(X_valid)
    print(estimator, mae(y_valid, predict))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
xgbclassifier = XGBClassifier(objective='multi:softmax', num_class=len(set(y_train)), n_estimators=700, learning_rate=0.01)
xgbclassifier.fit(X_train, y_train)

predict = xgbclassifier.predict(X_valid)
print(accuracy_score(y_valid, predict))
```
